# ndarray的创建

```python
import numpy as np
```

## 基础的创建

```python
arr = np.array([1,2,3]) #独自创建
list = [4.4,5.4,6.4]
arr2 = np.array(list) #利用列表创建
arr3 = np.array(list,dtype=np.int64) #创建时修改数据类型
print(arr)
print(arr2)
print(arr3)
```

```python
arr_cp = np.copy(arr) #利用copy创建 只复制值，是一个新的ndarray
print(arr_cp)
```

```python
#固定创建
arr_0 = np.zeros((2,3))
arr_1 = np.ones((3,4))
arr_empty = np.empty((5,6))
arr_full = np.full((7,8),2026)
print(arr_0)
print(arr_1)
print(arr_empty)
print(arr_full)
```

```python
#利用创建，但是只复制行列值，数值自己填
arr_like_0 = np.zeros_like(arr_full)
print(arr_like_0)
```
