# ndarray的特性
## 多维性

```python
import numpy as np
```

```python
arr1 = np.array(5)
arr2 = np.array([1,2,3])
arr3 = np.array([[1,2,3],[3,4,5]])
print('维度分别为：%d,%d,%d' % (arr1.ndim, arr2.ndim, arr3.ndim))
```

## 同质性

```python
arr = np.array([1,'hello'])
print(arr)
```

# ndarray的属性
## shape、ndim、size、dtype

```python
arr = np.array([[1,'2',3],[4,'你好','world']])
print('数组的行列',arr.shape)
print('数组的维度',arr.ndim)
print('数组的元素个数',arr.size)
print('数组的类型', 'string' if str(arr.dtype) == '<U21' else 'other')
```
